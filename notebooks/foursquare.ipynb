# Connecting to the Foursquare

### Note: 

This was a file where we first got familiar with the Foursqaure api, explored the JSON and got an idea of what columns we ultimately we wanted to extract. The dataset with which we ultimately work with for Foursquare is created in the 'bike_poi_radius_yelp.ipynb' file.

While it plays no direct purpose, it showcases part of the entire process.

```python
import os
import pandas as pd
import requests
from dotenv import load_dotenv
import json
```

```python
# loading the .env file
load_dotenv()
```

```python
# assigning stored api_key in .env file to vairables
api_key = os.environ.get('FOURSQUARE_API_KEY')
print(api_key)
```

```python
# building out the API, with criteria for bars (13003) and nightclubs (10032)
fsq_url = "https://api.foursquare.com/v3/places/search"

fsq_headers = {
	"Accept": "application/json",
    "Authorization": api_key
}

fsq_params = {
    "categories": '12080, 13003',
    "ll": '43.651070,-79.347015',
    "radius": 50000,
    "limit": 10
}

response = requests.request("GET", fsq_url, params=fsq_params, headers=fsq_headers)

data = response.json()
```

```python
# printing the json
print(json.dumps(data, indent=4))
```

```python
# Creating lists of columns ill want to have in final dataframe
fsq_id = []
ids = []
names = []
place_name = []
latitudes = []
longitudes = []

# pulling whats needed from 'results'
for results in data['results']:
    fsq = results['fsq_id']
    latitude = results['geocodes']['main']['latitude'] # Extracting latitudes and longitudes of places
    longitude = results['geocodes']['main']['longitude']
    
    # pulling whats needed from 'categories'
    categories = results['categories']
    for category in categories:
        _id = category['id'] # extracting ids of categories
        name = category['name'] # names of categories
        
        # appending to created lists above
        fsq_id.append(fsq)
        ids.append(_id)
        names.append(name)
        place_name.append(results['name'])
        latitudes.append(latitude)
        longitudes.append(longitude)


```

```python
data_dict = {
    'fsq_id': fsq_id,
    'category_id': ids,
    'category_name': names,
    'Place_Name': place_name,
    'latitude': latitudes,
    'longitude': longitudes
}
```

```python
bars_and_libs = pd.DataFrame(data_dict)
bars_and_libs
```
